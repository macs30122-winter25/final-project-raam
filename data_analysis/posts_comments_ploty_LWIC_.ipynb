```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## FOR ALL DATASETS

# Load the datasets
healthanxiety_posts_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/health_anxiety_posts_liwc.csv")
healthanxiety_comments_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/health_anxiety_comments_liwc.csv")

nursing_posts_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/nursing_posts_liwc.csv")
nursing_comments_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/nursing_comments_liwc.csv")

teaching_posts_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/teaching_posts_liwc.csv")
teaching_comments_df = pd.read_csv("C:/Users/amrit/GitHub/winterquarter/final-project-raam/Raam_database_csvs/teaching_comments_liwc.csv")

# Merge posts and comments on post_id
healthanxiety_merged_df = healthanxiety_comments_df.merge(healthanxiety_posts_df, on="post_id", suffixes=("_comment", "_post"))
nursing_merged_df = nursing_comments_df.merge(nursing_posts_df, on="post_id", suffixes=("_comment", "_post"))
teaching_merged_df = teaching_comments_df.merge(teaching_posts_df, on="post_id", suffixes=("_comment", "_post"))

# Select relevant columns
healthanxiety_merged_df = healthanxiety_merged_df[["post_id", "emo_anx_post", "emo_anx_comment"]]
nursing_merged_df = nursing_merged_df[["post_id", "emo_anx_post", "emo_anx_comment"]]
teaching_merged_df = teaching_merged_df[["post_id", "emo_anx_post", "emo_anx_comment"]]

# Combine all datasets
all_data = pd.concat([healthanxiety_merged_df, nursing_merged_df, teaching_merged_df], ignore_index=True)

# Rename columns for clarity
all_data.columns = ["Post ID", "Post Emo_Anx", "Comment Emo_Anx"]

# Display the first few rows
print(all_data.head())
```

```python
## PLOT ALL TOGETHER

# Plot using Seaborn Regression
plt.figure(figsize=(10, 6))
sns.regplot(x=all_data["Post Emo_Anx"], y=all_data["Comment Emo_Anx"], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel("Post Emo_Anx (Anxiety Score)")
plt.ylabel("Comment Emo_Anx (Anxiety Score)")
plt.title("Combined Emo_Anx Scores: Posts vs. Comments (Seaborn Regression)")
plt.grid(True)
plt.show()
```

```python
## PLOT LINE CHART

# Plot using Seaborn Line Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=all_data["Post Emo_Anx"], y=all_data["Comment Emo_Anx"], marker='o')
plt.xlabel("Post Emo_Anx (Anxiety Score)")
plt.ylabel("Comment Emo_Anx (Anxiety Score)")
plt.title("Combined Emo_Anx Scores: Posts vs. Comments (Seaborn Line Plot)")
plt.grid(True)
plt.show()
```
